### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingreso Tienda 1:", ingreso_tienda1)
print("Ingreso Tienda 2:", ingreso_tienda2)
print("Ingreso Tienda 3:", ingreso_tienda3)
print("Ingreso Tienda 4:", ingreso_tienda4)

```

# 2. Ventas por categoría

```python

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for nombre, df in zip(nombres, tiendas):
    print(f"\n {nombre} - Categorías más populares")
    ventas = df.groupby('Categoría del Producto').size().reset_index(name='cantidad_vendida')
    ventas = ventas.sort_values(by='cantidad_vendida', ascending=False)
    print(ventas)
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

print("Promedio Tienda 1:", promedio_tienda1)
print("Promedio Tienda 2:", promedio_tienda2)
print("Promedio Tienda 3:", promedio_tienda3)
print("Promedio Tienda 4:", promedio_tienda4)

```

# 4. Productos más y menos vendidos

```python
import pandas as pd


tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for nombre, df in zip(nombres, tiendas):
    print(f"\n {nombre}")


    ventas = df.groupby("Producto").size().reset_index(name="Cantidad Vendida")
    ventas = ventas.sort_values(by="Cantidad Vendida", ascending=False)


    mas_vendido = ventas.iloc[0]

    menos_vendido = ventas.iloc[-1]

    print("Más vendido:", mas_vendido["Producto"], "-", mas_vendido["Cantidad Vendida"])
    print("Menos vendido:", menos_vendido["Producto"], "-", menos_vendido["Cantidad Vendida"])
```

```python
import matplotlib.pyplot as plt

for nombre, df in zip(nombres, tiendas):
    ventas = df.groupby("Producto").size().reset_index(name="Cantidad Vendida")
    ventas = ventas.sort_values(by="Cantidad Vendida", ascending=False)

    plt.figure(figsize=(8,4))
    plt.bar(ventas["Producto"], ventas["Cantidad Vendida"], color="skyblue")
    plt.xticks(rotation=90, ha="right")
    plt.title(f"Ventas por producto - {nombre}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad Vendida")
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
promedio_envio1 = tienda["Costo de envío"].mean()
promedio_envio2 = tienda2["Costo de envío"].mean()
promedio_envio3 = tienda3["Costo de envío"].mean()
promedio_envio4 = tienda4["Costo de envío"].mean()

print("Costo envío promedio Tienda 1:", promedio_envio1)
print("Costo envío promedio Tienda 2:", promedio_envio2)
print("Costo envío promedio Tienda 3:", promedio_envio3)
print("Costo envío promedio Tienda 4:", promedio_envio4)
```

```python
import matplotlib.pyplot as plt

ingresos = {
    "Tienda 1": ingreso_tienda1,
    "Tienda 2": ingreso_tienda2,
    "Tienda 3": ingreso_tienda3,
    "Tienda 4": ingreso_tienda4,
}

plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color="teal")
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

ventas_categoria = tienda3.groupby('Categoría del Producto').size()

plt.figure(figsize=(6,6))
plt.pie(ventas_categoria, labels=ventas_categoria.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Ventas por Categoría - Tienda 1")
plt.show()
```

```python

calificaciones_promedio = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]
costos_envio_promedio = [promedio_envio1, promedio_envio2, promedio_envio3, promedio_envio4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
plt.scatter(costos_envio_promedio, calificaciones_promedio, color='purple')

for i, nombre in enumerate(tiendas):
    plt.text(costos_envio_promedio[i], calificaciones_promedio[i], nombre)

plt.title("Calificación Promedio vs. Costo de Envío Promedio por Tienda")
plt.xlabel("Costo de Envío Promedio")
plt.ylabel("Calificación Promedio")
plt.grid(True)
plt.show()
```
